```python
#import necessary libraries
import numpy as np
from PIL import Image
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense
```

```python
# #Loading image data
# Make sure they are at least 3 images
img1 = Image.open('cat1.jpeg').convert('L')
img2 = Image.open('dog1.jpeg').convert('L')
img3 = Image.open('cat3.jpeg').convert('L')

# Resize images to 28x28
resized_images = [
    np.array(img1.resize((28, 28))),
    np.array(img2.resize((28, 28))),
    np.array(img3.resize((28, 28)))
]
```

```python

x_data = np.array(resized_images).reshape(-1, 28, 28, 1) / 255.0
y_data = np.array([0, 1, 0])
```

```python
x_train, x_test = x_data[:2], x_data[2:]
y_train, y_test = y_data[:2], y_data[2:]
```

```python
model = Sequential()
model.add(Conv2D(16, (3, 3), strides=(1, 1), activation='relu', input_shape=(28, 28, 1)))
model.add(Conv2D(32, (3, 3), strides=(2, 2), activation='relu'))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```
